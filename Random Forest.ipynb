# Random Forest Algorithm
# We will be using it to classify the digits images from sklearn library

```python
# Loading the important libraries
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
dir(digits)
```

```python
# Plotting the digits to have a look using matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray() 
for i in range(4):
    plt.matshow(digits.images[i]) 
```

```python
# creating a dataframe from this dataset
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
```

```python
df[0:12]
```

```python
# Importing train test slpit to set the train and test dataset
X = df.drop('target',axis='columns')
y = df.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Importing the Random Forest classifier to train the model 
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
# score should be as high as possible
```

```python
y_predicted = model.predict(X_test)
```

```python
# We would plot a confusion matrix to check the model performance 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
